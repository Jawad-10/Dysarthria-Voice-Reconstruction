```python

import numpy as np
from sklearn.preprocessing import StandardScaler
import IPython.display as ipd
import pandas as pd
```

```python
data = pd.read_csv(r"D:\dys\Dysarthria and Non Dysarthria\torgo.csv")
dys_df = data[data['Is_dysarthria']=='Yes']
non_dys_df = data[data['Is_dysarthria']=='No']
dys_df.head()
```

```python
import torch
from torchvggish import vggish, vggish_input
```

```python
model = vggish()
model.eval()
```

```python
dys_files = dys_df['Wav_path'].tolist()
non_dys_files = non_dys_df['Wav_path'].tolist()
```

```python

import os

def extract_embedding_from_audio(audio_path, expected_features=384):
    try:
        # Normalize the file path
        audio_path = os.path.normpath(audio_path)
        
        # Perform feature extraction
        example = vggish_input.wavfile_to_examples(audio_path)
        
        # Check if feature extraction was successful
        if example.nelement() == 0:
            print(f"Failed to extract features from audio file: {audio_path}")
            return None
        
        # Perform model inference
        embeddings = model.forward(example)
        
        # Convert embeddings to NumPy array
        embeddings_np = embeddings.detach().numpy()
        
        print(f"Successfully extracted features from audio file: {audio_path}")
        return embeddings_np
    except Exception as e:
        print(f"Error processing file: {audio_path}")
        print(e)
        return None
```

```python
dys_embeddings = [extract_embedding_from_audio(file) for file in dys_files]
dys_embeddings = [emb for emb in dys_embeddings if emb is not None]  # filter out None values

non_dys_embeddings = [extract_embedding_from_audio(file) for file in non_dys_files]
non_dys_embeddings = [emb for emb in non_dys_embeddings if emb is not None]  # filter out None values
```

```python
import joblib
X = dys_embeddings + non_dys_embeddings


y = [1]*len(dys_embeddings) + [0]*len(non_dys_embeddings)
```

```python
print("Number of samples in X:", len(X))
print("Number of samples in y:", len(y))
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


```

```python
print("Number of samples in X_train:", len(X_train))
print("Number of samples in y_train:", len(y_train))
print("Number of samples in X_test:", len(X_test))
print("Number of samples in y_test:", len(y_test))
```

```python
consistent_length = len(X_train[0])
X_train, y_train = zip(*[(x, y) for x, y in zip(X_train, y_train) if len(x) == consistent_length])
X_test, y_test = zip(*[(x, y) for x, y in zip(X_test, y_test) if len(x) == consistent_length])
```

```python
# Convert the tuple to a NumPy array
X_train_array = np.array(X_train)
# Reshape the array to have two dimensions
X_train_2d = X_train_array.reshape(X_train_array.shape[0], -1)

# Check the shape of the reshaped array
print(X_train_2d.shape)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python

rf.fit(X_train_2d, y_train)
```

```python
# Convert the tuple to a NumPy array
X_test_array = np.array(X_test)

# Reshape the array to have two dimensions
X_test_2d = X_test_array.reshape(X_test_array.shape[0], -1)

# Make predictions
y_pred = rf.predict(X_test_2d)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=["Non-dys","dys"])
print("Accuracy", accuracy)
print("\n Classification report: \n", report)
```
